```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

#funcion auxiliar para cargar dataframe desde un CSV en Google Drive
def loadCSVFromGoogleDrive(url):
  file_id=url.split('/')[-2]
  dwn_url='https://drive.google.com/uc?id=' + file_id
  df = pd.read_csv(dwn_url) #cargamos dataframe
  return df
```

```python
df = loadCSVFromGoogleDrive("https://drive.google.com/file/d/1lR4U71co_hlySpMel79OqmMVJZTETdZf/view?usp=sharing")
print(df.values)

```

```python
X = df.loc[:, df.columns != '2do_Ataque_Corazon']
y = df['2do_Ataque_Corazon'].values
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.30, random_state=
0, shuffle=True)

```

```python
lr = LogisticRegression()
lr = lr.fit(train_X, train_y)
```

```python
y_pred = lr.predict(test_X)
print("Predicted vs Expected")
print(y_pred)
print(test_y)
```

```python
print(classification_report(test_y, y_pred, digits=3))
```

```python
confusion_matrix(test_y, y_pred)
```
